# 4.Group By

## 단일 집계 함수 적용

```python
#sql
pd.read_sql(f"""
SELECT sex, count(*)
FROM tips
"""
, engine)
```

```python
#sql
# group by 추가로 sex별 결과 확인 가능
pd.read_sql(f"""
SELECT sex, count(*)
FROM tips
GROUP BY sex;
"""
, engine)
```

size() : 해당 열의 레코드 수를 반환하며, 결측값도 포함한 결과 값

```python
#python
tips.groupby("sex").size()
```

count() 각 열에 함수를 적용하여 각 열의 레코드 수를 반환하며, 결측값을 제외한 결과 값

```python
#python
tips.groupby("sex").count()
```

count() 를 사용하여 하나의 열에도 적용이 가능함

```python
#python
tips.groupby("sex")["total_bill"].count()
```

## 한 번에 여러 집계 함수 적용

팁 금액이 요일별로 어떻게 다른지 확인

agg() 를 사용하여 그룹화된 DataFrame에 특정 열에 함수를 적용할 수 있음

```python
#sql
pd.read_sql(f"""
SELECT day, AVG(tip), COUNT(*)
FROM tips
GROUP BY day;
"""
, engine)
```

```python
#python
# agg로 각 열에 대해 함수 사용. np.mean으로 평균, np.size로 개수 구함
tips.groupby("day").agg({"tip": np.mean, "day": np.size})
```

## 둘 이상의 열을 기준으로 그룹화

```python
#sql
pd.read_sql(f"""
SELECT smoker, day, COUNT(*), AVG(tip)
FROM tips
GROUP BY smoker, day;
"""
, engine)
```

groupby() 메서드에 열 목록을 전달하여 적용함

```python
#python
# 두 개 이상이 들어갈 때는 list 처리해서 넣어야 함
tips.groupby(["smoker", "day"]).agg({"tip": [np.size, np.mean]})
```

# 5.JOIN

예제 데이터 생성

```python
# randn은 평균이 0이고 표준편차가 1인 가우시안 표준정규분포 값을 생성해주는 것
# randn(4)이므로 1x4 형태의 array로 나옴
df1 = pd.DataFrame({"key": ["A", "B", "C", "D"], "value": np.random.randn(4)})
df2 = pd.DataFrame({"key": ["B", "D", "D", "E"], "value": np.random.randn(4)})
df1.to_sql('df1', con=engine, index=False, if_exists='replace')
df2.to_sql('df2', con=engine, index=False, if_exists='replace')
```

```python
df1
```

```python
df2
```

merge() 메서드에는 수행할 조인 유형(LEFT, RIGHT, INNER, FULL) 또는 조인할 컬럼(열 이름 또는 인덱스)을 지정할 수 있는 매개변수가 존재함

## INNER JOIN

```python
#sql
# df1의 key와 df2의 key가 같은 부분을 찾아 묶어서 내보냄
pd.read_sql(f"""
SELECT *
FROM df1
INNER JOIN df2
  ON df1.key = df2.key;
"""
, engine)
```

```python
#python
# on="key"로 비교할 열 지정하고 하나의 dataframe으로 합침
pd.merge(df1, df2, on="key")
```

## LEFT JOIN

```python
#sql
# left join df2이므로 df1 기준으로 df2 붙임
pd.read_sql(f"""
SELECT *
FROM df1
LEFT OUTER JOIN df2
  ON df1.key = df2.key;
"""
, engine)
```

```python
#python
pd.merge(df1, df2, on="key", how="left")
```

## FULL JOIN

```python
#python
# 대충 합집합이라고 생각하면 됨
pd.merge(df1, df2, on="key", how="outer")
```

# 6.UNION

## UNION ALL

테이블을 아래위로 데이터 중복에 상관없이 붙일 수 있음

```python
df1 = pd.DataFrame(
    {"city": ["Chicago", "San Francisco", "New York City"], "rank": range(1, 4)}
)
df2 = pd.DataFrame(
    {"city": ["Chicago", "Boston", "Los Angeles"], "rank": [1, 4, 5]}
)
df1.to_sql('df1', con=engine, index=False, if_exists='replace')
df2.to_sql('df2', con=engine, index=False, if_exists='replace')
```

```python
df1
```

```python
df2
```

```python
#sql
# union all로 그냥 하나로 합쳐버림
pd.read_sql(f"""
SELECT city, rank
FROM df1

UNION ALL

SELECT city, rank
FROM df2;
"""
, engine)
```

```python
#python
# concatenate 진행
pd.concat([df1, df2])
```

## UNION

테이블을 아래위로 데이터 중복없이 붙일 수 있음

```python
#sql
# union all은 중복 허용. 그냥 union은 중복 없이.
pd.read_sql(f"""
SELECT city, rank
FROM df1

UNION

SELECT city, rank
FROM df2;
"""
, engine)
```

```python
#python
# drop_duplicates로 중복 제거
pd.concat([df1, df2]).drop_duplicates()
```

# 8.Ranking

## 그룹당 상위 n개 행 

SQL : ROW_NUMBER()

```python
#sql
pd.read_sql(f"""
SELECT * FROM (
  SELECT
    t.*,
    ROW_NUMBER() OVER(PARTITION BY day ORDER BY total_bill DESC) AS rn
  FROM tips t
)
WHERE rn < 3
ORDER BY day, rn;
""", engine)
```

pandas : rank(method='first')

```python
#python
(
    tips.assign(
        rnk=tips.groupby(["day"])["total_bill"].rank(
            method="first", ascending=False
        )
    )
    .query("rnk < 3")
    .sort_values(["day", "rnk"])
)
```

SQL : RANK()

```python
#sql
pd.read_sql(f"""
SELECT * FROM (
  SELECT
    t.*,
    RANK() OVER(PARTITION BY sex ORDER BY tip) AS rnk
  FROM tips t
  WHERE tip < 2
)
WHERE rnk < 3
ORDER BY sex, rnk;
"""
, engine)
```

pandas : rank(method='min')

```python
(
    tips[tips["tip"] < 2]
    .assign(rnk_min=tips.groupby(["sex"])["tip"].rank(method="min"))
    .query("rnk_min < 3")
    .sort_values(["sex", "rnk_min"])
)
```
